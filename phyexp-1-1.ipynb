```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

* 高さheight、幅widthの画像変数を生成する場合は、numpy(OpenCVが利用している数値演算モジュール)の関数を使う。
* import numpy as np
* image = np.zeros((height,width,3), np.uint8) #カラー画像(3ch)の場合

上のやり方で作ったimageは全画素が0の黒い画像になる。

* 画像imageの縦i, 横jの画素に値valをセットするときは、image.itemset((i,j), val) と書く。
* pyplotのimshow()は白黒のコントラストを自動調整するので、正確に表示するには、plt.imshow(img,cmap="gray",vmin=0,vmax=255) と書く必要がある。


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
height=150
width=150

img = np.zeros((height,width), np.uint8)

for i in range(height):
    for j in range(width):
        I=min(i+j,255)
        img.itemset((i,j),I)
        
    
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
```

## 第１課題ファイルの最終課題

次の動作をするプログラムを作成せよ。

1. ファイルから画像を読み込み表示する。
2. R,G,Bに対応するトラックバーをつくり、それぞれのトラックバーを動かしてR,G,Bの成分の強さの倍率を変化させた画像を画面に表示する。

### 考察
倍率が大きくなると元の画像とは異なる色になってしまうのを観察せよ。なぜそのようなことが起こるのか理由を考えてレポートに示せ。またこれを回避するプログラムコードを考えて示せ。

### ヒント
* 読み込んだ画像とは別に画像を保存する変数を用意する必要がある。
* 読み込んだ画像の各画素の値にトラックバーの値(0~255)を掛け算し100で割ると、倍率(0~2.55)を掛けた画像になる。これをR,G,Bのチャンネルごとに行う。
* すべて画素ごとに上記の掛け算をするpython(numpy)の表現方法は？
* for文で画素アクセスしてもよいが、もっと速い書き方がチュートリアルに記載されている。

```python
import cv2
import numpy as np

def nothing(x):
    pass

image1=cv2.imread("lena.png")
image2=cv2.imread("lena.png")
img = np.zeros((image1.shape[0],image1.shape[1],3), np.uint8)
cv2.namedWindow('image')




cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)


switch = '0 : OFF \n1 : ON'
cv2.createTrackbar(switch, 'image',0,1,nothing)


while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
        
    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G','image')
    b = cv2.getTrackbarPos('B','image')
    s = cv2.getTrackbarPos(switch,'image')

    B,G,R=cv2.split(image1)
    B=image1[:,:,0].astype(float)
    G=image1[:,:,1].astype(float)
    R=image1[:,:,2].astype(float)
   
    B2=B*b/100
    G2=G*g/100
    R2=R*r/100

    
    image2[:,:,0] =np.where(B2>255,255,B2.astype(np.uint8))
    image2[:,:,1] =np.where(G2>255,255,G2.astype(np.uint8))
    image2[:,:,2] =np.where(R2>255,255,R2.astype(np.uint8))

    if s == 0:
        img[:] = 0
    else:
        img=image2
        

cv2.destroyAllWindows()
```

## imshowウインドウが消えなくなったときのおまじない

destroyAllWindows()関数で消去できるが、この関数を実行しただけでは反応しない。OpenCVのhighguiモジュール（ウィンドウやマウス・キーボード関係のインターフェースモジュール）はwaitKey()関数がコールされたときにアップデートを実行する（それまではイベントキューに溜まっているだけ）。
以下のコマンドでも消えないときは、jupyter kernelを再起動する。このタブの一番先頭にあるメニューのKernel中のRestartでリセットされる。この場合それまでに計算したり読み込んだ変数はすべて消える。読み込んだモジュールも初期化されるので、importコマンドなどは実行し直す必要がある。

```python
cv2.destroyAllWindows()
cv2.waitKey(1)
```
